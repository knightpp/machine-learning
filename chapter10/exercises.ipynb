## Exercise 10.1

$$ f_1(x_1, x_2) = sigmoid(x_1*1 + x_2*-2 + 1*-1) $$
$$ f_2(x_1, x_2) = sigmoid(x_1*-1 + x_2*3 + 1*-1) $$
$$ f_3(x_1, x_2) = sigmoid(f_1(x_1, x_2)*-1 + f_2(x_1, x_2)*2 + 1*1) $$


```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def f1(x1, x2):
    return sigmoid(x1+x2*-2-1)
def f2(x1, x2):
    return sigmoid(x1*-1+x2*3-1)
def f3(x1, x2):
    return sigmoid(f1(x1, x2)*-1 + f2(x1, x2)*2 + 1)

f3(1, 1)
```

## Exercise 10.2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras import activations
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

```python
data = {
    'x1': [0, 0, 1, 1],
    'x2': [0, 1, 0, 1],
    'y': [0, 1, 1, 0]
}
df = pd.DataFrame(data=data)
x = np.array(df[['x1', 'x2']])
y = np.array(to_categorical(df['y'], 2))
df
```

```python
model = Sequential()
model.add(Dense(2, activation=activations.relu, input_shape=(2,)))
model.add(Dense(8, activation=activations.relu))
model.add(Dense(2, activation=activations.softmax))
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
model.summary()
```

```python
# use 1000+ epochs to achieve accuracy 1.0
model.fit(x, y, epochs=200)
```
