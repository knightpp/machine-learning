```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../chapter3/Houses.csv', index_col=0)
df
```

```python
global_id = 0
city_name_to_id = {}
cat_cities = []

for (_, name) in enumerate(df['city']):
    if name in city_name_to_id:
        cat_cities.append(city_name_to_id[name])
    else:
        city_name_to_id[name] = global_id
        global_id += 1
        cat_cities.append(city_name_to_id[name])
```

```python
x = df.drop(['address', 'city', 'id', 'sq', 'latitude', 'longitude'], axis=1)
x['city'] = cat_cities
x
```

```python
y = df['price']
y
```

```python
model = Sequential()
model.add(Dense(5, activation="relu", input_shape=(5,)))
model.add(Dropout(.1))
model.add(Dense(128, activation="relu"))
model.add(Dropout(.1))
model.add(Dense(128, activation="relu"))
model.add(Dropout(.1))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer="adam",
              metrics=[keras.metrics.RootMeanSquaredError()])
model.summary()
```

```python
model.fit(x, y, epochs=10, batch_size=10)
```
