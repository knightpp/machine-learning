```python
from matplotlib import pyplot as plt
import numpy as np
import utils
import linear
```

# Defining and plotting dataset

```python
features = np.array([1, 2, 3, 5, 6, 7])
labels = np.array([155, 197, 244, 356, 407, 448])
# plt.figure(dpi=150)
utils.plot_points(features, labels)
```

```python
model = linear.regression(features, labels, epochs=1_000)
```

```python
print(f"absolute error      = {model.absolute_error(features, labels)}")
print(f"square error        = {model.square_error(features, labels)}")
print(f"mean absolute error = {model.mean_absolute_error(features, labels)}")
print(f"rsme                = {model.root_mean_square_error(features, labels)}")
```
